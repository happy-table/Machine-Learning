```
import numpy as np
import pandas as pd
bike = pd.read_csv("bike_shairing_hourly.csv")
print(bike)
print(bike.head())
```

```
bike.iloc[12]
```

```
bike.iloc[13,5]
```

```
print(bike.season.unique())
print(bike.yr.unique())
print(bike.mnth.unique())
print(bike.hr.unique())
print(bike.dteday.unique())
```

```
print(bike.isnull())
print(bike.isnull().sum())
print(bike.isna())
print("Duplicate records",bike.duplicated())
```

```
dte = pd.to_datetime(bike.dteday)
print(dte)
print(type(dte))
year = dte.dt.year
month = dte.dt.month
hour = dte.dt.hour
print("Years :",year.head(2))
print(year.tail(2))
print("Months are :", month )
print("Hours are :", hour )
```

```
bike.drop(["yr"], axis=1 , inplace=True)
bike.columns

```

```

df_by_seasons = bike.groupby ("season")
print(type(df_by_seasons))
print(df_by_seasons["casual"].sum())
print(df_by_seasons["registered"].sum())
```
