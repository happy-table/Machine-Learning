```python
import matplotlib.pyplot as mp
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
type(tips)
```

```python
(tips.columns)
```

```python
tips.head()
```

```python
tips.tail()
```

```python
sns.scatterplot(data = tips , x=tips["sex"], y=tips["tip"] )
mp.title("Tip",size=20 , pad = 20 , color = "red")
```

```python
sns.scatterplot(data = tips , x="total_bill", y="tip" )
mp.title("Tip",size=20 , pad = 20 , color = "red")
```

```python
sns.lineplot(data = tips , x="total_bill", y="tip" )
mp.title("Tip",size=20 , pad = 20 , color = "red")
```

```python
sns.relplot(data = tips , x="total_bill", y="tip")
```

```python
sns.relplot(data = tips , x="total_bill", y="tip",kind="line")
```

```python
sns.relplot(data = tips , x="total_bill", y="tip",hue="time")
```

```python
sns.relplot(data = tips , x="total_bill", y="tip",hue="time",kind="line")
```
