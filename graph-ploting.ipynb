```python
x = np.array([5,6,7,8])
y = np.arange(len(x))
mp.plot(x,y,"o-.")
mp.grid(True)
#adding label to graph
mp.xlabel("Average Height")
mp.ylabel("Class")
mp.title("Average height per class")
mp.show()
```

```python
x = np.array([5,6,7,8])
y = np.arange(len(x))

font1= {"family":"serif", "color":"red", "size" : 20}
font2= {"family":"serif", "color":"green", "size" : 10}

mp.plot(x,y,"o-.")
mp.grid(True)
#adding label to graph
mp.xlabel("Average Height", fontdict=font2)
mp.ylabel("Class",fontdict=font2)
mp.title("Average height per class",fontdict=font1,pad=20)
mp.show()
```

```python
bikes = pd.read_csv("bike_shairing_hourly.csv")
bikes.head()
```

```python
x=bikes.season
mp.plot(x,"o--")
mp.grid(True)
```

```python
mp.plot(bikes.temp,"-",linewidth="0.5")
mp.grid(True)
```

```python
x = bikes.casual
mp.plot(x,"-",linewidth=0.25)
```

```python
s = bikes.groupby("season")
x = s["casual"].sum()
mp.plot(x,"o-.")
```

```python
s = bikes.groupby("season")
mp.plot(bikes.season.unique(),s.cnt.sum(),"o-.")
mp.ticklabel_format(style="plain")
mp.grid(1)
```
