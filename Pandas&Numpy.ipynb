```
import numpy as np
import pandas as pd
list1=[i*i for i in range(5)]
array = np.random.randint(5,size=5)
dic={
    "Column1" : array,
    "Column2" : list1,
}
print(dic)
df = pd.DataFrame(dic)
print(df)
```

```
list_of_indices = ["index1","index2","index3","index4","index5"]
s1 = pd.Series([1,2,3,4,5],index=list_of_indices)
s2 = pd.Series([11,22,33,44,55],index=list_of_indices)
s3 = pd.Series(["A","B","C","D","E"],index=list_of_indices)
list_of_series = [s1,s2,s3]
df = pd.concat(list_of_series, axis=1)
print(df)
```
