```
Income= k['INCOME'].values
Spend = k['SPEND'].values
x = np.array(list(zip(Income,Spend)))
#x= np.cloumn_stack((Income,Spend))
x
```

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters=i,random_state=0)
  km.fit(x)
  wcss.append(km.inertia_)

plt.plot(range(1,11),wcss,color='red',marker='d')
plt.title('Optimal k value')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.grid()
plt.show()
```

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters=5,random_state=0)
  km.fit(x)
  wcss.append(km.inertia_)

plt.plot(range(1,11),wcss,color='red',marker='d')
plt.title('Optimal k value')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.grid()
plt.show()
```

#**Fitting the k-means to dataset with k=4**


```
model = KMeans(n_clusters=4,random_state=0)
y_mean =model.fit_predict(x)
y_mean
```


####Cluster 1: Customers with medium income and low spend
####Cluster 2: Customers with high income and medium to high spend
####Cluster 3: Customers with low income
####Cluster 4: Customers with medium income but high spend

#**Scatter Plot**

```

plt.scatter(x[y_mean==0,0],x[y_mean==0,1],s=70,c='red',label='Cluster 1')
plt.scatter(x[y_mean==1,0],x[y_mean==1,1],s=70,c='blue',label='Cluster 2')
plt.scatter(x[y_mean==2,0],x[y_mean==2,1],s=70,c='green',label='Cluster 3')
plt.scatter(x[y_mean==3,0],x[y_mean==3,1],s=70,c='cyan',label='Cluster 4')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=200,c='yellow',label='Centroids', marker='s')
plt.title('Income Spent Analysis')
plt.xlabel('Income')
plt.ylabel('Spend')
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

# Number of clusters (automatically detected)
n_clusters = len(np.unique(y_mean))

# Generate a colormap for distinct colors
colors = cm.rainbow(np.linspace(0, 1, n_clusters))

# Plot each cluster with loop (cleaner, scalable)
for i, color in zip(range(n_clusters), colors):
    plt.scatter(
        x[y_mean == i, 0],
        x[y_mean == i, 1],
        s=70,
        color=color,
        label=f'Cluster {i + 1}'
    )

# Plot centroids
plt.scatter(
    model.cluster_centers_[:, 0],
    model.cluster_centers_[:, 1],
    s=200,
    c='black',
    label='Centroids',
    marker='X'
)

# Optional: Annotate each centroid with its index
for idx, (cx, cy) in enumerate(model.cluster_centers_):
    plt.text(cx, cy, f'C{idx+1}', fontsize=12, fontweight='bold', color='orange',
             ha='center', va='center')

# Titles and labels
plt.title('Income vs Spend Cluster Analysis')
plt.xlabel('Income')
plt.ylabel('Spend')
plt.legend()
plt.tight_layout()
plt.show()
```
