# Q4.Plot the distribution of all numerical features and the categorical target using matplotlib and observe the plots.

```python
# ans
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
df = sns.load_dataset("iris")

# Plot distribution of numerical features
df.hist(figsize=(6, 4), edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.tight_layout()
plt.show()

# Plot categorical target (species) count
df['species'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Count of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

# Q5.Plot a "featurepair-wise" scatter plot to see how the numerical features are correlated to each other and print out pair-wise correlation coefficients between the numerical features .

```python
# ans
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('iris.csv')

# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# 1. Pairwise scatter plot
sns.pairplot(numeric_df)
plt.show()

# Correlation matrix
correlation_matrix = numeric_df.corr()
print("Pairwise Correlation Coefficients:")
print(correlation_matrix)
```
